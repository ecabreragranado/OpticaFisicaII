## Teora

Normalmente el experimento de Young se representa con una doble rendija, tal y como aparece en la figura anterior. Una onda esfrica (o bien una onda plana, el tratamiento es equivalente), incide en una pantalla sobre la cual se han realizado dos aperturas $S_1$ y $S_2$ muy prximas entre s (llamaremos a la distancia entre ellas $a$). Estas aperturas actan como dos fuentes secundarias de radiacin, generando a su vez dos ondas esfricas que se superponen en el espacio que hay detrs de ellas. Si observamos la distribucin de irradiancia en una pantalla situada a una cierta distancia $D$, qu nos encontraremos?.

Las dos ondas que se generan en $S_1$ y $S_2$ pueden escribirse como:

$$\vec{E_1} = \vec{e_1} \; E_{01} \;  \; \cos\left( k r_1 - \omega t + \phi_1\right)$$
$$\vec{E_2} = \vec{e_2} \; E_{02} \;  \; \cos\left( k r_2 - \omega t + \phi_2\right)$$

$E_{0j}$  es la amplitud de la onda, $\vec{e_j}$  es la direccin de vibracin y $\phi_j$ es la fase inicial.  $r_1$ ($r_2$) es el camino que recorre la onda desde $S_1$ ($S_2$) hasta el punto de observacin P. Ambas ondas tienen la misma longitud de onda.

La superposicin de estas dos ondas, nos dar la expresin de la irradiancia ya conocida, 

$$I_T = I_1 + I_2 + 2 \sqrt{I_1 I_2} \; (\vec{e_ 1}\cdot\vec{e_ 2}) \; cos(\delta)$$

donde $\delta = k (r_2 - r_1) + \phi_2 - \phi_1$ es el desfase (o la diferencia de fase) entre las dos ondas.

En esta expresin podemos hacer alguna que otra simplificacin, 

* $\vec{e_ 1} \cdot \vec{e_ 2}=1$ porque las ondas las consideramos polarizadas linealmente en la misma direccin.
* $I_1 = I_2$ en caso de que no haya ningn filtro en $S_1$  $S_2$, las dos ondas tienen la misma amplitud.
* $\phi_2 - \phi_1 = 0$ ya que el frente de ondas llega simultneamente a $S_1$ y $S_2$. Ntese que si colocsemos por ejemplo una pieza de un material transparente antes de una de las dos aperturas, tendramos un desfase adicional en una de las dos ondas y esta diferencia ya no sera nula. Esto ocurrira porque una de las ondas viajara a travs del material mientras la otra onda lo hara en aire. 

As la irradiancia total queda

$$I_T = 2 I_1 \left( 1 + cos(\delta) \; \right)$$

con $\delta = k (r_2 - r_1)$

Como vemos, **es la diferencia de caminos $\Delta = r_2 - r_1$ la que determina el valor de la irradiancia final en el punto P**. Vamos a calcularla.

```python
from IPython.display import Image
Image(filename="ExperimentoYoung.jpg")
```

Segn la figura, $\Delta = r_2 - r_1$ lo podemos escribir como $\Delta = a sen(\theta)$, siendo $a$ la separacin entre las rendijas. Si ste ngulo es pequeo (lo que significa que la distancia entre las fuentes y la pantalla de observacin sea grande comparada con la separacin entre las fuentes), esta expresin la podemos simplificar, 

$$ \Delta = a sen(\theta) \simeq a tan(\theta) = a \frac{x}{D}$$.

Y por tanto, 

$$\delta = k \frac{a x }{D} = \frac{2 \pi a x}{\lambda D}$$

En estas expresiones, $x$ es la distancia del punto P de observacin al eje mientras que $D$ es la distancia entre el plano
que contiene a las fuentes y la pantalla de observacin, donde se encuentra P. Podemos reescribir la irradiancia total en la pantalla empleando la expresin calculada del desfase 

$$I_T = 2 I_1 \left( 1 + cos\left( \frac{2 \pi a x}{\lambda D} \right) \; \right)$$

## Distribucin de luz. Patrn de interferencias

Ahora estamos en disposicin de contestar a la pregunta que nos plantebamos antes, cmo es la distribucin de irradiancia en la pantalla de observacin?. Vemos que el desfase depende de la altura en la pantalla $x$, por tanto al movernos en esa direccin el valor de la irradiancia cambiar. En particular el trmino que provoca esa variacin es del tipo cosenoidal $cos( \frac{2 \pi a x}{\lambda D})$ por lo que veremos en la pantalla una distribucin cosenoidal, con mximos de irradiancia cuando $\delta = 2 m \pi$, con $m = 0, \pm 1, \pm 2 ...$ y mnimos de irradiancia cuando $\delta = (2 m + 1) \pi$, con $m = 0, \pm 1, \pm 2 ...$. Las posiciones $x$ a las que corresponden estas condiciones sern, 

Mximos de irradiancia.  $\delta = 2 m \pi \implies \Delta = m \lambda \implies$
$$x^{max}_m = \frac{m \lambda D}{a}$$

Mnimos de irradiancia.  $\delta = (2 m + 1)\pi \implies \Delta = \frac{(2m +1) \lambda}{2} \implies$
$$x^{min}_m = \frac{(2m + 1) \lambda D}{2a}$$

Vamos a dibujar la distribucin de irradiancia en la pantalla y un corte a lo largo del eje X (ejecutar la siguiente celda de cdigo).

```python
from matplotlib.pyplot import *
from numpy import *
%matplotlib inline
style.use('fivethirtyeight')

###################################################################################
# PARMETROS. SE PUEDEN MODIFICAR SUS VALORES
###################################################################################
Lambda =400e-9 # en metros, longitud de onda de la radiacin
D = 4.5 # en metros, distancia entre el plano que contiene las fuentes y la pantalla de observacin
a = 0.003 # en metros, separacin entre fuentes
###################################################################################

interfranja=Lambda*D/a # clculo de la interfranja
k = 2.0*pi/Lambda
x = linspace(-5*interfranja,5*interfranja,500)
I1 = 1 # Consideramos irradiancias normalizadas a un cierto valor.
I2 = 0.01
X,Y = meshgrid(x,x)
delta = k*a*X/D
Itotal = I1 + I2 + 2.0*sqrt(I1*I2)*cos(delta)
figure(figsize=(14,5))
subplot(121)
pcolormesh(x*1e3,x*1e3,Itotal,cmap = 'gray',vmin=0,vmax=4)
xlabel("x (mm)"); ylabel("y (mm)")
subplot(122)
plot(x*1e3,Itotal[x.shape[0]/2,:])
xlabel("x (mm)"); ylabel("Irradiancia total normalizada")
```

Como podemos ver, los mximos estn equiespaciados (lo mismo sucede con los mminos), siendo la distancia entre dos mximos consecutivos
$$ \text{Interfranja} = \frac{\lambda D}{a} $$
Dicha magnitud se conoce con el nombre de interfranja y nos da informacin sobre el tamao caracterstico del patrn de franjas. 
Adems del tamao, para poder observar con claridad las franjas es necesario que estn bien contrastadas. Para ello se define el contraste o visibilidad de las franjas
$$ C = \frac{I_T^{max}-I_T^{min}}{I_T^{max}+I_T^{min}}$$
que nos dice cuanto estn separados los mximos de luz respecto de los mnimos. 

El valor de estas dos magnitudes para el caso representado en la figura anterior se muestra en la siguiente celda (ejecutar dicha celda despus de haber ejecutado la anterior celda de cdigo)

```python
interfranja=Lambda*D/a # clculo de la interfranja
C = (Itotal.max() - Itotal.min())/(Itotal.max() + Itotal.min()) # clculo del contraste
print "a=",a*1e3,"mm   ","D=",D,"m   ","Longitud de onda=",Lambda*1e9,"nm" # valores de los parmetros
print "Interfranja=",interfranja*1e3,"mm" # muestra el valor de la interfranja en mm
print 'Contraste=',C # muestra el valor del contraste
```
